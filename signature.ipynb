```python
def divh(img,h,k):
    if(k==7):
        return img
    else: 
        (X, Y) = img.shape
        m = np.zeros((X, Y))

        for x in range(X):
            for y in range(Y):
                m[x, y] = img[x][y]!= 255
                
        m = m / np.sum(np.sum(m))

        dx = np.sum(m, 1)
        dy = np.sum(m, 0)

        cx = np.sum(dx * np.arange(X))
        cy = np.sum(dy * np.arange(Y))
        c.append([cx,cy])
        if(h==1):
            img3=img[:,int(cy):]
            img4=img[:,:int(cy)]
            img3=divh(img3,0,k+1)
            img4=divh(img4,0,k+1)
        else:
            img3=img[int(cx):,:]
            img4=img[:int(cx),:]
            img3=divh(img3,1,k+1)
            img4=divh(img4,1,k+1)

        if(h==0):
            img5=np.concatenate((img4, img3),axis=0)
            for i in range(Y):
                img5[int(cx)][i]=0
        else:
            img5=np.concatenate((img4, img3),axis=1)
            for i in range(X):
                img5[i][int(cy)]=0
        return img5
```
